```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import time
from matplotlib import pyplot as plt
from datetime import datetime
from datetime import timedelta
```

```python
filename = 'pulls.csv'
df = pd.read_csv(filename)
df = df.drop(df.columns[[0]], axis=1)
df['Seconds'] = pd.to_timedelta(df['Duration']).astype('timedelta64[s]').astype(int)
```

```python
p1s = df[df['Instance'] == 0]
p1s = p1s.sort_values(by='Seconds')
p1s = p1s.reset_index(drop=True)

p1s['Event'] = ''
p1s.loc[(p1s['Event'] == '') & (p1s['Seconds'] < 116), 'Event'] = 'Aetherial Shackles + Flails'
p1s.loc[(p1s['Event'] == '') & (p1s['Seconds'] < 196), 'Event'] = 'Intemperance #1'
p1s.loc[(p1s['Event'] == '') & (p1s['Seconds'] < 291), 'Event'] = 'Aetherflails'
p1s.loc[(p1s['Event'] == '') & (p1s['Seconds'] < 339), 'Event'] = 'Fourfold Shackles'
p1s.loc[(p1s['Event'] == '') & (p1s['Seconds'] < 413), 'Event'] = 'Intemperance #2'
p1s.loc[(p1s['Event'] == '') & (p1s['Seconds'] < 554), 'Event'] = 'Aetherflails'
p1s.loc[(p1s['Event'] == ''), 'Event'] = 'Warder\'s Wraths'
p1s.loc[(p1s['Clear'] == True), 'Event'] = 'Clear'
```

```python
mpl.style.use('seaborn')
fig, ax = plt.subplots(figsize=(30,20))
for event in p1s['Event'].unique():
    group = p1s[p1s['Event'] == event]
    ax.barh(group.index, group['Seconds'], label=f'{event} ({len(group)} | {100 * len(group) / len(p1s):.1f}%)')
max_index = p1s.index.max()
max_sec = p1s['Seconds'].max()
x_ticks_sec = np.arange(0, max_sec, 60)
x_ticks = [time.strftime("%M:%S", time.gmtime(x)) for x in x_ticks_sec]
ax.set_xticks(x_ticks_sec, x_ticks)
ax.set_xlim(0, max_sec)
ax.set_ylim(0, max_index)
ax.invert_yaxis()
ax.set_title(f'P1S ({len(p1s)} Pulls)', fontsize=64)
ax.set_xlabel('Pull Duration', fontsize=36)
ax.set_ylabel('Pull', fontsize=36)
ax.tick_params(axis='both', which='major', labelsize=24)
ax.legend(fontsize=36)
```
